```python
df_long_word_dup1 = df_long_word.drop_duplicates(
    subset=["lForm", "lemma"], keep="first"
)
df_long_word_dup2 = df_long_word_dup1.drop_duplicates(subset=["lemma"], keep="first")
df_long_word_dup2
```

```python
df_merge_pron_lemma = pd.merge(
    df_pron_not_in_base[["word_id", "spell", "frequency"]],
    df_long_word_dup2[["lForm", "lemma"]],
    left_on="spell",
    right_on="lemma",
    how="left",
    indicator="check",
)
to_csv_sig(
    df_merge_pron_lemma,
    "D:/Dropbox/06.wanjuan/99.tmp/df_merge_pron_lemma.csv",
)
l("ok")
```

```python
df_merge_pron_lemma["long_word_pron_hira"] = df_merge_pron_lemma["lForm"].apply(
    lambda x: jaconv.kata2hira(x) if pd.notnull(x) and x != "" else x
)

kks = pykakasi.kakasi()


def to_hira(text):
    text_list = kks.convert(text)
    return "".join([item["hira"] for item in text_list])


df_merge_pron_lemma["remain_pron_hira"] = df_merge_pron_lemma["spell"].apply(
    lambda x: to_hira(x) if pd.notnull(x) and x != "" else x
)

df_merge_pron_lemma["pron"] = df_merge_pron_lemma["long_word_pron_hira"].fillna(
    df_merge_pron_lemma["remain_pron_hira"]
)

katsu = cutlet.Cutlet()
df_merge_pron_lemma["roman"] = df_merge_pron_lemma["pron"].apply(katsu.romaji)


to_csv_sig(
    df_merge_pron_lemma,
    "D:/Dropbox/06.wanjuan/99.tmp/df_merge_pron_lemma.csv",
)
l("ok")
```

```python
df_merge_pron_lemma['pron_id']=df_merge_pron_lemma['word_id'].astype(str)+'10'
df_merge_pron_lemma['pron_percent']=1
df_merge_pron_lemma['pron_index']=1
del df_merge_pron_lemma['remain_pron_hira']
del df_merge_pron_lemma['long_word_pron_hira']
del df_merge_pron_lemma['check']
del df_merge_pron_lemma['lemma']
del df_merge_pron_lemma['lForm']
del df_merge_pron_lemma['frequency']


to_csv_sig(
    df_merge_pron_lemma,
    "D:/Dropbox/06.wanjuan/99.tmp/df_merge_pron_lemma.csv",
)
l("ok")
```
