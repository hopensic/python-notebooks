```python
df_long_word_dup1 = df_long_word.drop_duplicates(subset=["lForm", "lemma"],
                                                 keep="first")
df_long_word_dup2 = df_long_word_dup1.drop_duplicates(subset=["lemma"],
                                                      keep="first")
df_long_word_dup2
```

```python
df_merge_pron_lemma = pd.merge(
    df_pron_not_in_base[["word_id", "spell", "frequency"]],
    df_long_word_dup2[["lForm", "lemma"]],
    left_on="spell",
    right_on="lemma",
    how="left",
    indicator="check",
)
to_csv_sig(
    df_merge_pron_lemma,
    "D:/Dropbox/06.wanjuan/99.tmp/df_merge_pron_lemma.csv",
)
l("ok")
```

```python
df_merge_pron_lemma["long_word_pron_hira"] = df_merge_pron_lemma["lForm"].apply(
    lambda x: jaconv.kata2hira(x) if pd.notnull(x) and x != "" else x)

kks = pykakasi.kakasi()


def to_hira(text):
    text_list = kks.convert(text)
    return "".join([item["hira"] for item in text_list])


df_merge_pron_lemma["remain_pron_hira"] = df_merge_pron_lemma["spell"].apply(
    lambda x: to_hira(x) if pd.notnull(x) and x != "" else x)

df_merge_pron_lemma["pron"] = df_merge_pron_lemma["long_word_pron_hira"].fillna(
    df_merge_pron_lemma["remain_pron_hira"])

katsu = cutlet.Cutlet()
df_merge_pron_lemma["roman"] = df_merge_pron_lemma["pron"].apply(katsu.romaji)

to_csv_sig(
    df_merge_pron_lemma,
    "D:/Dropbox/06.wanjuan/99.tmp/df_merge_pron_lemma.csv",
)
l("ok")
```

```python
df_merge_pron_lemma['pron_id'] = df_merge_pron_lemma['word_id'].astype(
    str) + '10'
df_merge_pron_lemma['pron_percent'] = 1
df_merge_pron_lemma['pron_index'] = 1
del df_merge_pron_lemma['remain_pron_hira']
del df_merge_pron_lemma['long_word_pron_hira']
del df_merge_pron_lemma['check']
del df_merge_pron_lemma['lemma']
del df_merge_pron_lemma['lForm']
del df_merge_pron_lemma['frequency']

to_csv_sig(
    df_merge_pron_lemma,
    "D:/Dropbox/06.wanjuan/99.tmp/df_merge_pron_lemma.csv",
)
l("ok")
```

```python
# 关联词性ID
df_incr_meaning_merged_v3["pos_id"] = (df_incr_meaning_merged_v3["pos"].map(
    df_pos_dict.set_index("pos")["pos_id"]).astype("Int64"))

# 关联发音数据集
df_incr_meaning_merged_v3_idx = df_incr_meaning_merged_v3.set_index(
    ["spell", "pron"])
df_new_pron_idx = df_new_pron_135405_with_freq_not_in_meaning_lt_5000.set_index(
    ["spell", "pron"])

df_merged_1 = df_incr_meaning_merged_v3_idx.merge(
    # df_hira_new[["pron_id", "mix_pron_freq"]],
    df_new_pron_idx[["pron_id"]],
    left_index=True,
    right_index=True,
    how="left",
    indicator="check_pron",
)
to_csv_sig(df_merged_1, "D:/Dropbox/06.wanjuan/99.tmp/df_merged_1.csv", True)
l("ok")

replace_dict = {"both": "normal", "left_only": "pron_missing"}
df_merged_1["check_pron"] = df_merged_1["check_pron"].map(replace_dict)
df_merged_1 = df_merged_1.reset_index()

# 关联基础词频数据集
df_merged_1.set_index(["spell"], inplace=True)
df_merged_raw_word_agg_idx = df_merged_raw_word_agg.set_index(["spell"])
df_merged_2 = df_merged_1.merge(
    df_merged_raw_word_agg_idx[["word_id"]],
    left_index=True,
    right_index=True,
    how="left",
    indicator="check_spell",
)
# 字符串替换，把拼写找不到的标识一下
replace_dict_2 = {"both": "normal", "left_only": "spell_missing"}
df_merged_2["check_spell"] = df_merged_2["check_spell"].map(replace_dict_2)
df_merged_2 = df_merged_2.reset_index()
to_csv_sig(df_merged_2, "D:/Dropbox/06.wanjuan/99.tmp/df_merged_2.csv", True)
l("ok")
```

```python
# 将单行的词义转为多行的词义
extracted = df_merged_2["meaning_desc"].str.extractall(r"\(([^:)]+):([^)]+)%\)")
extracted = extracted.reset_index(level=1, drop=True).rename(
    columns={0: "meaning", 1: "meaning_per"}
)
extracted["meaning_per"] = extracted["meaning_per"].astype(float)
result = df_merged_2.drop(columns=["meaning_desc"]).join(extracted)

result["pron_per"] = round(result["pron_per"] / 100, 2)
result["pos_per"] = round(result["pos_per"] / 100, 2)
result["meaning_per"] = round(result["meaning_per"] / 100, 2)

result["meaning_per"] = round(
    result["pron_per"] * result["pos_per"] * result["meaning_per"], 2
)

column_order = [
    "meaning_id",
    "spell",
    "pron",
    "pos",
    "meaning",
    "meaning_per",
    "pos_per",
    "pron_per",
    "word_id",
    "pron_id",
    "check_spell",
    "check_pron",
]
result = result.reset_index(drop=True)
start_index = 5000
result["meaning_id"] = range(start_index, start_index + len(result))
result = result[column_order]

to_csv_sig(result, "D:/Dropbox/06.wanjuan/99.tmp/normal_merged_result_0729.csv")
l("ok")
```
