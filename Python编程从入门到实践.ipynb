## 3.2增删改列表中的元素

````python
a=[1,2,3]
# 直接修改
a[1]='99'
# 通过append来向末尾增加元素
a.append(4)
# 通过insert向列表的某个位置增加元素
a.insert(1,'first')
print('a='+str(a))

#列表中删除元素
b=[4,5,6]
#del 的方式删除
del b[1]
print('b='+str(b))

#pop 的方式删除，是从尾部删除。
#pop可以返回被删除的元素
#pop也可以弹出任意位置的元素
c=[7,8,9,10,11]
c.pop()
print('c='+str(c))
poped = c.pop()
print('poped='+str(poped))
d=[7,8,9,10,11]
print('poped='+str(d.pop(2)))


#remove 删除遇到的第一个值
e=[1,2,3,1,5]
e.remove(1)
print('e='+str(e))
````

## 4.3 创建数字列表

````python
#使用range()来创建数字
a = range(1,5)
for x in a:
    print(x)
print(a)

#可以用list()将range转换为列表
b=list(range(1,3))
print(b)

#range还可以抱定初始定和步长
c=list(range(2,10,3))
print(c)

#列表解析
squares=[value**2 for value in range(1,6)]
print(squares)
    
    
````

# Ch8. 函数

## 8.2 传递实参

````python
# 位置实参

def test(types, name):
    print('types:'+types)
    print('name:'+name)
test('aType','aName')
print('-----------------------------')

# 关键字实参
test(name='bName',types='bType')
print('-----------------------------')

# 函数可以有默认值
def test2(types,name='defaultName'):
    print('types:'+types)
    print('name:'+name)
test2('cType')
test2('cType','cName')

print('-----------------------------')
````

## 8.4 传递列表

````python
# 在函数中修改列表
def changeList(p_list):
    p_list[0]=99
a=[1,2,3]
changeList(a)
print(a)
print('----------------------')

# 禁止函数修改列表(传递副本)
def forbidChangeList(p_list):
    p_list[0]=88
b=[4,5,6]
forbidChangeList(b[:])
print(b)

````
