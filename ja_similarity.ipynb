```python
import os
import pickle
import pandas as pd
import jp_util.common_path as cp
from jp_util.pandas_util import rd_csv_sig, to_csv_sig
from jp_util.file_util import get_obj_by_pickle_path, export_to_pickle_path
from jp_util.common_util import l, p
import jp_util.jap_base_class as jp
from importlib import reload
```

```python
# df_similarity_path = (
#     "D:/02.work_project/jap_base_structure/similarity/pkl/df_similarity_5000.pkl"
# )

df_similarity_path = (
    "D:/02.work_project/jap_base_structure/similarity/pkl/df_similarity_5000_v2.pkl"
)

def get_similarity_df():
    if os.path.exists(df_similarity_path):        
        with open(df_similarity_path, "rb") as f:
            return pickle.load(f)


df_similarity = get_similarity_df()
# df_base = rd_csv_sig("D:/Dropbox/06.wanjuan/99.tmp/df_increased_3000_final_v3.csv")
df_base = rd_csv_sig('d:/tmp/df_base_word_remain.csv')

df_similarity_disturb = pd.merge(
    df_similarity,
    df_base[["word_id", "spell", "frequency", "roman", "simple_meaning"]],
    how="left",
    left_on="disturb_word_id",
    right_on="word_id",
)


del df_similarity_disturb["word_id_y"]
df_similarity_disturb.rename(
    columns={
        "spell": "disturb_word_spell",
        "frequency": "disturb_word_frequency",
        "roman": "disturb_word_roman",
        "simple_meaning": "disturb_word_meaning",
    },
    inplace=True,
)


to_csv_sig(df_similarity_disturb, "d:/tmp/df_similarity_disturb_v2.csv")


df_disturb_base = pd.merge(
    df_similarity_disturb,
    df_base[["word_id", "spell", "frequency", "roman", "simple_meaning"]],
    how="left",
    left_on="word_id_x",
    right_on="word_id",
)

del df_disturb_base["word_id"]
df_disturb_base.rename(
    columns={"word_id_x": "word_id"},
    inplace=True,
)


df_disturb_base["disturb_pid"] = range(1, 1 + len(df_disturb_base))

df_disturb_base_order = [
    "disturb_pid",
    "disturb_word_id",
    "disturb_word_spell",
    "disturb_word_roman",
    "disturb_word_frequency",
    "disturb_word_meaning",
    "similarity",
    "word_id",
    "spell",
    "frequency",
    "simple_meaning",
    "roman",
]

df_disturb_base = df_disturb_base[df_disturb_base_order]
df_disturb_base.sort_values(
    ["frequency", "similarity", "disturb_word_frequency"],
    ascending=[True, False, True],
    inplace=True,
)
to_csv_sig(df_disturb_base, "d:/tmp/df_disturb_base_zzz_v2.csv")
l('ok')
```
